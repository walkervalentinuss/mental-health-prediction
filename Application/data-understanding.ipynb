## Missing Value

```python
def missing_data(df):
    """
    Menghitung jumlah total dan persentase nilai yang hilang (missing values) di setiap kolom pada DataFrame.
    
    Parameters:
    df : pd.DataFrame
        DataFrame yang akan diperiksa missing values-nya.
    
    Returns:
    pd.DataFrame
        DataFrame yang berisi jumlah dan persentase missing values di setiap kolom.
    """
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)
    
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Missing Percent'])
    missing_data['Missing Percent'] = missing_data['Missing Percent'].apply(lambda x: x * 100)
    
    return missing_data
```

# Data Preprocessing

## Handling Missing Value Categorical

```python
cat_treshold = 100
```

```python
for col in train_df.select_dtypes('object').columns:
    train_df[col] = train_df[col].fillna('missing')
    train_df.loc[train_df[col].value_counts(dropna=False)[train_df[col]].values < cat_treshold, col] = 'noise'
```

```python
for col in test_df.select_dtypes('object').columns:
    test_df[col] = test_df[col].fillna('missing')
    test_df.loc[test_df[col].value_counts(dropna=False)[test_df[col]].values < cat_treshold, col] = 'noise'
```

```python
missing_data(train_df)
```

```python
missing_data(test_df)
```

```python
num_cols_train = [cols for cols in train_df.select_dtypes('number').columns if cols!= 'id' and cols!='Depression']
```

```python
for col in num_cols_train:
    train_df[col]=train_df[col].fillna(0)
```

```python
train_df.isnull().sum()
```

```python
num_cols_test = [cols for cols in test_df.select_dtypes('number').columns if cols!= 'id']
```

```python
for col in num_cols_test:
    test_df[col]=test_df[col].fillna(0)
```

```python
train_df['Work/Academic Pressure']=train_df['Work Pressure']+train_df['Academic Pressure']
test_df['Work/Academic Pressure']=test_df['Work Pressure']+test_df['Academic Pressure']
```

## Mapping Sleep Duration

```python
train_df['Sleep Duration'].value_counts()
```

```python
def sleep_duration(row):
    if(row['Sleep Duration']=='Less than 5 hours'):
        return 4
    elif (row['Sleep Duration']=='7-8 hours'):
        return 7.5
    elif (row['Sleep Duration']=='More than 8 hours'):
        return 9
    elif (row['Sleep Duration']=='5-6 hours'):
        return 5.5
    else:
        return 7
```

```python
train_df['Sleep Duration'] = train_df.apply(sleep_duration, axis=1)
```

```python
test_df['Sleep Duration'] = test_df.apply(sleep_duration, axis=1)
```

```python
train_df = train_df[train_df['Dietary Habits'] != 'noise']
test_df = test_df[test_df['Dietary Habits'] != 'noise']
```

```python
# Fill 0 with median for Work/Study Hours (median is eligible for handling outlier)
train_df['Work/Study Hours'] = train_df['Work/Study Hours'].replace(0, train_df['Work/Study Hours'].median())

# Fill 0 with mode for Financial Stress (there are 4 values so we use mode to keep another value dominance)
train_df['Financial Stress'] = train_df['Financial Stress'].replace(0, train_df['Financial Stress'].mode()[0])

# Fill 0 with mode for Job/Study Satisfaction
train_df['Job/Study Satisfaction'] = train_df['Job/Study Satisfaction'].replace(0, train_df['Job/Study Satisfaction'].mode()[0])

# Fill 0 with mode for Work/Academic Pressure
train_df['Work/Academic Pressure'] = train_df['Work/Academic Pressure'].replace(0, train_df['Work/Academic Pressure'].mode()[0])
```

```python
# Fill 0 with median for Work/Study Hours (median is eligible for handling outlier)
test_df['Work/Study Hours'] = test_df['Work/Study Hours'].replace(0, test_df['Work/Study Hours'].median())

# Fill 0 with mode for Financial Stress (there are 4 values so we use mode to keep another value dominance)
test_df['Financial Stress'] = test_df['Financial Stress'].replace(0, test_df['Financial Stress'].mode()[0])

# Fill 0 with mode for Job/Study Satisfaction
test_df['Job/Study Satisfaction'] = test_df['Job/Study Satisfaction'].replace(0, test_df['Job/Study Satisfaction'].mode()[0])

# Fill 0 with mode for Work/Academic Pressure
test_df['Work/Academic Pressure'] = test_df['Work/Academic Pressure'].replace(0, test_df['Work/Academic Pressure'].mode()[0])
```

```python
for col in train_df.columns:
    print(f"Value counts for {col}:")
    print(train_df[col].value_counts())
    print("\n")
```

## Encoding

```python
label_encoder = LabelEncoder()
ordinal_encoder = OrdinalEncoder()
```

```python
binary_columns = ['Gender', 'Working Professional or Student', 'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']
ordinal_columns = ['Dietary Habits', 'Financial Stress']
```

```python
for col in binary_columns:
    train_df[col] = label_encoder.fit_transform(train_df[col])
```

```python
for col in ordinal_columns:
    train_df[col] = ordinal_encoder.fit_transform(train_df[[col]])
```

```python
for col in binary_columns:
    test_df[col] = label_encoder.fit_transform(test_df[col])
```

```python
for col in ordinal_columns:
    test_df[col] = ordinal_encoder.fit_transform(test_df[[col]])
```
